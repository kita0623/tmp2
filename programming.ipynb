# 辞書

```python
# ディクショナリの作成
my_dict = {'apple': 100, 'banana': 50, 'orange': 80}

# キーを指定して値を取得
print(my_dict['apple'])  # 100

# 新しいキーと値のペアを追加
my_dict['grape'] = 120

# キーと値のペアを更新
my_dict['banana'] = 70

# キーが存在するかどうかをチェック
if 'orange' in my_dict:
    print('orange is in my_dict')
else:
    print('orange is not in my_dict')

# キーのリストを取得
keys = list(my_dict.keys())
print(keys)  # ['apple', 'banana', 'orange', 'grape']

# 値のリストを取得
values = list(my_dict.values())
print(values)  # [100, 70, 80, 120]

# キーと値のペアのリストを取得
items = list(my_dict.items())
print(items)  # [('apple', 100), ('banana', 70), ('orange', 80), ('grape', 120)]

# キーを指定して値を削除
del my_dict['orange']

# ディクショナリをループして、すべてのキーと値を取得
for key, value in my_dict.items():
    print(key, value)
```

# クラス

```python
# クラスの定義
class Triangle:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

# クラスの使用
triangle1 = Triangle(10, 5)
print(triangle1.area())

triangle2 = Triangle(6, 8)
print(triangle2.area())
```

```python
# 問題：
# あるECサイトの注文データがあるとします。注文データは、以下のようなディクショナリのリストとして与えられます。

orders = [
    {"order_id": 1, "item_name": "pen", "item_price": 100, "quantity": 2},
    {"order_id": 2, "item_name": "notebook", "item_price": 200, "quantity": 3},
    {"order_id": 3, "item_name": "ruler", "item_price": 50, "quantity": 1},
    {"order_id": 4, "item_name": "pencil", "item_price": 50, "quantity": 5},
]
```

# 演習：総合問題

```python
# 【問題】
# 以下のようなリストがあるとします。

data = [5, 3, 2, 1, 4, 6, 7, 8, 10, 9]

# このリストから以下の処理を順番に行って、結果を表示してください。

# リストの最小値を求める
# リストの最大値を求める
# リストの平均値を求める
# リストの中央値を求める（ヒント：リストを昇順にソートする必要があります）
# リストの要素を2倍にして新しいリストを作成する
```

```python
data = [5, 3, 2, 1, 4, 6, 7, 8, 10, 9]

# リストの最小値を求める
min_value = data[0]
for i in data:
    if i < min_value:
        min_value = i
print("最小値：", min_value)

# リストの最大値を求める
max_value = data[0]
for i in data:
    if i > max_value:
        max_value = i
print("最大値：", max_value)

# リストの平均値を求める
sum_value = 0
for i in data:
    sum_value += i
mean_value = sum_value / len(data)
print("平均値：", mean_value)

# リストの中央値を求める
sorted_data = sorted(data)
n = len(sorted_data)
if n % 2 == 0:
    median_value = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
else:
    median_value = sorted_data[n // 2]
print("中央値：", median_value)

# リストの要素を2倍にして新しいリストを作成する
new_data = []
for i in data:
    new_data.append(i * 2)
print("新しいリスト：", new_data)
```

シグモイド関数の数式
シグモイド関数の数式は以下のように表されます。

$$ f(x) = \frac{1}{1 + e^{-x}} $$

$x$は入力信号を表し、$f(x)$は0から1までの値を取る出力信号です。

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = sigmoid(x)

sns.set_style("whitegrid")
sns.lineplot(x=x, y=y)
plt.title("Sigmoid Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

# 3次元のデータを生成
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# 3次元グラフの描画
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, cmap="coolwarm")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
# 以下に、算術演算子と論理演算子と比較演算子を組み合わせた例を示します。

a = 10
b = 5
c = 20

result = ((a + b) > c) or ((a * b) > c)

print(result)

# この例では、変数aに10、変数bに5、変数cに20を代入しています。式(a + b) > cはFalseとなるため、次の式(a * b) < cが評価されます。
# 式(a * b)は50となり、cの値20より小さいため、式(a * b) < cはTrueとなります。このため、式((a + b) > c) or ((a * b) < c)の評価結果はTrueとなります。
# このように、複数の算術演算子、論理演算子、比較演算子を組み合わせることで、複雑な式を表現することができます。
```
