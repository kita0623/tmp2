```python
# 対数変換

# サンプルデータの作成
data = np.random.lognormal(4, 0.8, 1000)

# 対数変換前のヒストグラム
sns.histplot(data, kde=False)
plt.title('Histogram of original data')
plt.show()

# 対数変換後のヒストグラム
log_data = np.log(data)
sns.histplot(log_data, kde=False)
plt.title('Histogram of log-transformed data')
plt.show()
```

```python
import numpy as np
import seaborn as sns

# サンプルデータの作成
data = np.random.normal(10, 2, 1000)

# 正規化
normalized_data = (data - np.mean(data)) / np.std(data)

# ヒストグラムのプロット
sns.histplot(data, color='blue', alpha=0.5, bins=30)
sns.histplot(normalized_data, color='green', alpha=0.5, bins=30)

# # X軸とY軸の範囲を統一
# plt.xlim(-5, 25)
# plt.ylim(0, 0.25)

# グラフの表示
plt.show()
```

```python
import numpy as np
import seaborn as sns

# サンプルデータの作成
data = np.random.normal(10, 2, 1000)

# 標準化
standardized_data = (data - np.mean(data)) / np.std(data)

# グラフの分割とプロット
fig, axes = plt.subplots(ncols=2, figsize=(10, 5))
sns.histplot(data, ax=axes[0], color='blue', alpha=0.5, bins=30)
sns.histplot(standardized_data, ax=axes[1], color='green', alpha=0.5, bins=30)

# # X軸とY軸の範囲を統一
axes[0].set_xlim(0, 20)
# axes[0].set_ylim(0, 0.25)
axes[1].set_xlim(-5, 25)
# axes[1].set_ylim(0, 0.8)

# グラフの表示
plt.show()
```
