```python
# 対数変換

# サンプルデータの作成
data = np.random.lognormal(4, 0.8, 1000)

# 対数変換前のヒストグラム
sns.histplot(data, kde=False)
plt.title('Histogram of original data')
plt.show()

# 対数変換後のヒストグラム
log_data = np.log(data)
sns.histplot(log_data, kde=False)
plt.title('Histogram of log-transformed data')
plt.show()
```

```python
import pandas as pd
import random

fruits = ["apple", "banana", "orange", "grape", "mango"]
sales = [random.randint(1000, 5000) for _ in range(5)]

df = pd.DataFrame({"fruits": fruits, "sales": sales})

df = df.sort_values('sales', ascending=False)


# グラフの大きさを指定する
plt.figure(figsize=(6,6))

# 円グラフを描画する
# sns.set_palette("pastel")
# sns.set(font_scale=1.2)
plt.pie(df["sales"], labels=df["fruits"], autopct="%1.1f%%", startangle=90, counterclock=False)
# plt.pie(df["sales"], labels=df["fruits"], autopct="%1.1f%%", startangle=90)

# タイトルを追加する
plt.title("Fruit Sales")

# 円グラフを表示する
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# データの生成
labels = ['Sales', 'Marketing', 'HR', 'IT', 'Finance']
# labels = ['営業', 'マーケティング', '人事', 'IT', '財務']
sizes = [25, 15, 10, 20, 30]
colors = ['#6b9a3b', '#8ebd58', '#b7ce85', '#d2e2b9', '#beded0']

df = pd.DataFrame({'Department': labels, 'Budget(%)': sizes})
df_sorted = df.sort_values('Budget(%)', ascending=True)

# 円グラフの描画
plt.figure(figsize=(6,6))
plt.pie(df_sorted['Budget(%)'], labels=df_sorted['Department'], colors=colors[::-1],
        autopct='%1.1f%%', startangle=90, pctdistance=0.7, labeldistance=1.05)
plt.title('Annual Budget Allocation', fontsize=16, fontweight='bold')
# plt.title('年間予算配分', fontsize=16, fontweight='bold')
plt.axis('equal')
plt.show()
```

```python
# サンプルデータの生成
np.random.seed(42)
page_load_time = np.random.normal(loc=2.5, scale=1.0, size=10000)

# ヒストグラムの描画
plt.figure(figsize=(10, 5)) # グラフのサイズを変更
sns.histplot(page_load_time, kde=False, color='green', alpha=0.8)

# グラフの装飾
plt.title('ページ表示時間の分布')
plt.xlabel('ページ表示時間 (秒)')
plt.ylabel('アクセス数')

plt.show()
```

```python

# サンプルデータの生成
np.random.seed(123)
heights = np.random.normal(loc=170, scale=5, size=100)
weights = 0.7 * heights - 50 + np.random.normal(scale=5, size=100)

# データをPandasのデータフレームに変換
df = pd.DataFrame({'Height': heights, 'Weight': weights})

# Seabornを使用して散布図を描く
# sns.set(style="darkgrid")
plt.figure(figsize=(10, 5)) # グラフのサイズを変更
sns.scatterplot(data=df, x='Height', y='Weight', color='b', alpha=1)

# グラフにタイトル、X軸ラベル、Y軸ラベル、凡例を追加
plt.title('身長と体重の関係')
plt.xlabel('身長 (cm)')
plt.ylabel('体重 (kg)')
# plt.legend(loc='lower right')
plt.show()
```

```python

# サンプルデータの生成
np.random.seed(123)
sales = np.random.normal(loc=10, scale=0.5, size=12)
months = pd.date_range(start='2022-01-01', end='2022-12-01', freq='MS')
months = [month.strftime('%b') for month in months]

# データをPandasのデータフレームに変換
df = pd.DataFrame({'Month': months, 'Sales': sales})

# Seabornを使用して線グラフを描く
plt.figure(figsize=(10, 5)) # グラフのサイズを変更
sns.lineplot(data=df, x='Month', y='Sales', color='b')

# グラフにタイトル、X軸ラベル、Y軸ラベル、凡例を追加
plt.title('月間売上')
plt.xlabel('月')
plt.ylabel('売上 (億円)')
# plt.legend(labels=['Sales'], loc='upper left')
plt.show()
```
